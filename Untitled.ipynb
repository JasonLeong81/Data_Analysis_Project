```python
new = df.iloc[:,1:]
new.head()
# new.columns
# year = 2020 - year 
new['current_year'] = 2021
new['num_year'] = new['current_year'] -  new['Year']
new.drop(['current_year','Year'],axis=1,inplace=True) # axis 1 (column) axis 0 (rows)
new.head()
new = pd.get_dummies(new,drop_first=True) # dummy variable coding # 0 0 means the third one is active 
new.head()
```

```python
new.corr()
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
# sns.pairplot(new)
corrmat = new.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
g = sns.heatmap(new[top_corr_features].corr(),annot=True,cmap='RdYlGn') # darker color means positively correlated and lighter means negative
```

```python
new.head()
# independent and dependent features 
x = new.iloc[:,1:] # everything except for the first column 
y = new.iloc[:,0] # first column 
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(x,y)
print(model.feature_importances_) # check for which features are more important 

```
